We need to convert in this format -> ['Action', 'Adventure', 'Fantasy', 'SciFi']

```python
import ast   # for conversion of only string list to list with integer indices
def convert(obj):
  L = []
  for i in ast.literal_eval(obj):
    L.append(i['name'])
  return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies['keywords']= movies['keywords'].apply(convert)
```

```python
movies['keywords'].head(1)
```

```python
movies['cast'][0]
```

```python
def convert3(obj):
  L=[]
  counter = 0
  for i in ast.literal_eval(obj):
    if counter!=3:
      L.append(i['name'])
      counter+=1
    else:
      break
  return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head(2)
```

```python
def get_director(obj):
  L = []
  for i in ast.literal_eval(obj):
    if i['job'] == 'Director':
       L.append(i['name'])
       break
  return L

```

```python
movies['crew'] = movies['crew'].apply(get_director)
```

```python
movies.head(2)
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ", "") for i in x])
```

Now we will perform text vectorization where each word will be a vector in a vector space. We will select most frequent 5 thousand words from the entire tags and if there is five thousand movies in our dataset, we will count the frequency for each columns from word1 to word5000.

Each row will be a vector like this. the size of the vectors will be (5000,5000)

And then finding only similarity of them we can recommend the closest movie based on a random movie.

For doing this job we need to remove stop words like to, the, a etc which have no significance in meaning.

We need to convert all the words to root words. Like from dancing, danced, dance I want dance everywhere. This will help me to pick only unique 5000 words list. Not the same word again and again!

For this we will use a library. And this process is called stemming in text processing.

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
  y = []
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(df['tags']).toarray()
vectors
```

```python
# We can see all the feature names
print(len(cv.get_feature_names_out()))
cv.get_feature_names_out()
```

We will measure the similariry of movies using cosine similarity. We won't use euclidean distance since it fails to measure the distance of vectors correctly.
We will measure the distance among movies using angle using cosine similarity.

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
def recommend(movie):
  movie_index = df[df['title']==movie].index[0]
  distances = similarity[movie_index]
  recommended = sorted(list(enumerate(distances)), reverse=True, key = lambda x:x[1])[1:6]
  for i in recommended:
    print(df.iloc[i[0]]['title'])
```

```python
recommend('Batman Begins')
```
